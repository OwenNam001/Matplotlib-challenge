```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single DataFrame
mouse_study_results = pd.merge(study_results, mouse_metadata, on=["Mouse ID", "Mouse ID"], how="left")
mouse_study_results.count()
```

```python
mouse_study_results["Mouse ID"].nunique()
```

```python
grouped = study_results.groupby(["Mouse ID","Timepoint"]).size()
grouped[grouped > 1].reset_index(name="Count")
```

```python
mouse_study_results.loc[(mouse_study_results["Mouse ID"] == "g989")]
```

```python
mouse_study_results_clean = mouse_study_results.loc[(mouse_study_results["Mouse ID"] != "g989")]
mouse_study_results_clean
```

```python
len(mouse_study_results_clean["Mouse ID"].unique())
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

Mean_Tumor_Volume = mouse_study_results.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
Median_Tumor_Volume = mouse_study_results.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
Tumor_Volume_Variance = mouse_study_results.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
Tumor_Volume_Std_Dev = mouse_study_results.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
Tumor_Volume_Std_Err = mouse_study_results.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

Study_Summary_Statistics = pd.DataFrame(
                            {
                            "Mean Tumor Volume" : Mean_Tumor_Volume,
                            "Median Tumor Volume" : Median_Tumor_Volume,                                
                            "Tumor Volume Variance" : Tumor_Volume_Variance,
                            "Tumor Volume Std. Dev." : Tumor_Volume_Std_Dev,
                            "Tumor Volume Std. Err." : Tumor_Volume_Std_Err
                            })
Study_Summary_Statistics
```

```python
summary_statistics = mouse_study_results.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_statistics
```

```python
import numpy as np

study_count_df = mouse_study_results_clean.groupby(["Drug Regimen"])[["Mouse ID", "Timepoint"]].size() #.reset_index(name="Number of rows"))
study_count_sorted_df = study_count_df.sort_values(ascending=False)
study_count_sorted_df.plot(kind="bar", legend=False)

plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")

plt.show()
```

```python
import numpy as np

# Generate Pandas Series for the number of rows (Mouse ID/Timepoints) for each drug regimen
study_count = mouse_study_results_clean.groupby(["Drug Regimen"])[["Mouse ID", "Timepoint"]].size()

# Sorting the Number of rows with descending order
study_count = study_count.sort_values(ascending=False)

# set x_axis, y_axis
x_axis = study_count.index
y_axis = study_count.values

# Generate a bar chart using pyplot
plt.figure(figsize=(8,5))
plt.bar(x_axis, y_axis, alpha=1, align="center")

# set x_axis headers
plt.xticks(rotation="vertical")

# set x_axis, y_axis limitation
plt.xlim(-0.9, len(x_axis)-0.1)
plt.ylim(0, max(y_axis)+10)

# set x_axis, y_axis label
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")

plt.show()
```

```python
# Generate Pandas Series for the number of rows with each mouse sex
mouse_sex = mouse_study_results_clean.groupby(["Sex"]).size()

# Sorting the Number of rows with descending order
mouse_sex = mouse_sex.sort_values(ascending=False)

# Generate a pie chart with DataFrame.plot()
mouse_sex.plot(kind='pie', y=mouse_sex.values, autopct='%1.1f%%', startangle=0, legend=False)

# set ylabel = "Sex"
plt.ylabel("Sex")

plt.show()
```

```python
# Generate new dataFrame for "Drug Regimen" = "Capomulin" & "Mouse ID" = "l509"
l509_study = mouse_study_results_clean.loc[(mouse_study_results_clean["Drug Regimen"] == "Capomulin") & (mouse_study_results_clean["Mouse ID"] == "l509"), ["Timepoint","Tumor Volume (mm3)"]]

# set x_axis, y_axis
x_axis = l509_study["Timepoint"]
y_axis = l509_study["Tumor Volume (mm3)"]

plt.plot(x_axis, y_axis)

# set chart labels and a title
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Generate a new dataFrame for "Capomulin" Regimen
capomulin_data = mouse_study_results_clean.loc[(mouse_study_results_clean["Drug Regimen"] == "Capomulin"), ["Mouse ID", "Tumor Volume (mm3)", "Weight (g)"]]

# Calculate average tumor volume for each ["Weight (g)", "Mouse ID"]
capomulin_avg_tumor = capomulin_data.groupby(["Weight (g)", "Mouse ID"])["Tumor Volume (mm3)"].mean().reset_index(name="avg(Tumor)")

# set x_axis with Weight (g)
x_axis = capomulin_avg_tumor["Weight (g)"]

# set y_axis with average observed tumor volume
data = capomulin_avg_tumor["avg(Tumor)"]

# Generate a scatter plot with pyplot
plt.scatter(x_axis, data, marker="o", s=x_axis*2, alpha=1)

# set chart labels and a title
plt.xlabel("Weight (g) ")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```
